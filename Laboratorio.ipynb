```python
def f(x):return x**5-6*x**3+2*x-2
def df(x):return 5*x**4-18*x**2-2
```

```python
#ejemplo (lo que sucede abajo es que itera este procedimiento)
x=2.2
x=x-f(x)/df(x)
x
x=x-f(x)/df(x)
x
```

```python
import numpy as np
#Funcion y su derivada
fx  = lambda x: x**5-6*x**3+2*x-2
dfx = lambda x: 5*x**4-18*x**2+2

x0 = 2.2
tolera = 0.0001

tabla = []
Etramo = abs(2*tolera)
xi = x0
while (Etramo>=tolera):
    xnuevo = xi - fx(xi)/dfx(xi)
    Etramo  = abs(xnuevo-xi)
    tabla.append([xi,xnuevo,Etramo])
    xi = xnuevo

tabla = np.array(tabla)
n = len(tabla)

#la diferencia es el resultado de la resta entre el valor actual y el valor de x en la iteracion anterior xnuevo -xi
print(['xi', 'xnuevo', 'diferencia'])
np.set_printoptions(precision = 4)
print(tabla)
print('raiz en: ', xi)
print('con error de: ',Etramo)
```
